```python
from random import choice
import numpy as np
```

```python
class Perceptron: 
    
    def __init__(self):
        pass
       
    def train(self, X , Y):
        X = np.asarray(X)
        Y = np.asarray(Y)
        
        (n,m) = X.shape
        X = np.append(X, np.ones((n,1)),1)
        w = np.zeros(m+1)
        
        erro = 1
        maxiter = 100
        ciclo = 1
        while erro > 0 and ciclo < maxiter:
            erro = 0
            for y,x in zip(Y,X):
                dot = np.dot(w,x)
                if(dot <= 0):
                    yp = 0
                else:
                    yp = 1
                print("%a inner %a dot %i =>  %i = %i "%(w,x,dot,yp, y))
                if yp != y:
                    w = w + 1*(yp-y)*x
                    erro+=1
            ciclo+=1
            print(ciclo)
        
        return Predictor(w)
    
```

```python
class Predictor:
    
    def __init__(self, w):
        self.w = w
        pass
    
    def predict(self,X):   
        X = np.asarray(X)
        (n,m) = X.shape
        X = np.append(X, np.ones((n,1)),1)
        
        print(self.w)
        y=[]
        
        y = np.dot(X, self.w)
        
        return y
```

```python
x = [[0,0],
    [0,1],
    [1,0],
    [1,1]]

y = [0,1,1,1]

p = Perceptron();
preditor = p.train(x,y)
print(preditor.predict(x))
```
